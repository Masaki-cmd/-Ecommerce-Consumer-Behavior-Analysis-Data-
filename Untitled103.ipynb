### 主キー候補評価

データセット内で一意な行を識別するための主キー候補を評価します。主キーは、データ結合や分析において各エンティティを特定するために重要です。

**評価基準:**
*   一意性: その列の値がすべて異なるか。
*   非NULL性: その列に欠損値がないか。

（注意：現状では列名が不明なため、一般的な候補と評価のロジックを提示します。データフレーム読み込み後、実際の列名に基づいて評価を更新する必要があります。）

```python
if df is not None:
    print('主キー候補評価:')
    pk_candidates = []
    for col in df.columns:
        is_unique = df[col].is_unique
        has_no_null = df[col].isnull().sum() == 0

        reason = []
        if not is_unique: reason.append('一意性がありません。')
        if not has_no_null: reason.append('欠損値があります。')
        if is_unique and has_no_null: reason.append('主キー候補です。')
        if not reason: reason.append('主キー候補ではありません。')

        print(f'- 列「{col}」: {', '.join(reason)}')
        if is_unique and has_no_null:
            pk_candidates.append(col)

    if pk_candidates:
        print(f'\n主な主キー候補: {', '.join(pk_candidates)}')
    else:
        print('\n単一列での主キー候補は見つかりませんでした。複数列の組み合わせで検討が必要です。')
else:
    print('データフレームが読み込まれていないため、主キー候補評価を表示できません。')
```

## KGIの定義

### Subtask:
データから測定可能な、ビジネス目標の中心となるKGI（Key Goal Indicator）を1つ定義します。


### KGIの定義

**KGI (Key Goal Indicator):** 顧客生涯価値（Customer Lifetime Value: CLTV）の最大化

**選定理由:**

最終的な目標が「顧客ロイヤルティと売上収益の最大化に貢献するデータ駆動型の意思決定を可能にすること」であるため、顧客生涯価値（CLTV）の最大化をKGIとして設定します。

CLTVは、一顧客が企業にもたらす生涯にわたる総利益を測定するものであり、顧客ロイヤルティの向上はCLTVの増加に直結します。また、CLTVの最大化は長期的な売上収益の最大化に不可欠です。

このKGIは、以下のデータセットの列と関連付けることができます。

*   `Total_Purchase_Amount`: 個々の購入金額から顧客ごとの総購入金額を算出する基礎となります。
*   `Previous_Purchases`: 顧客の購入履歴やリピート購入の傾向を把握し、将来の購買行動を予測するために重要です。
*   `Loyalty_Program_Member`: ロイヤルティプログラムへの参加は、顧客のロイヤルティレベルと将来的なCLTVへの貢献度を示す指標となります。
*   `Customer_Rating`, `Customer_Reviews`: 顧客満足度はロイヤルティに強く影響し、CLTVの変動要因となります。

これらの情報を組み合わせることで、顧客ロイヤルティを測り、各顧客が将来的にどれだけの価値をもたらすかを推定し、CLTVを最大化するための戦略を策定することが可能となります。

## KPIの定義

### Subtask:
KGI達成のために追跡すべき、最大5つのKPI（Key Performance Indicator）を定義します。これらのKPIは、データセット内の列から直接測定可能である必要があります。KPIの選定理由も併記します。


### KPIの定義

KGIである**顧客生涯価値（CLTV）の最大化**を達成するために追跡すべき、以下の5つのKPIを定義します。

1.  **平均注文額 (Average Order Value - AOV)**
    *   **選定理由**: 一回の購入あたりの金額が増加することは、顧客が企業にもたらす総収益を直接的に増加させ、結果としてCLTVの向上に貢献します。
    *   **測定可能列**: `Average_Order_Value` (この列が直接利用可能でない場合、`Total_Purchase_Amount`と`Frequency_of_Purchases`から顧客ごとの平均を算出できますが、ここでは直接の列として存在すると仮定します。)

2.  **購入頻度 (Purchase Frequency)**
    *   **選定理由**: 顧客がより頻繁に購入することは、顧客とのエンゲージメントの高さとリピート購買行動を示し、生涯にわたる総購入回数を増やすことでCLTVを高めます。
    *   **測定可能列**: `Frequency_of_Purchases`

3.  **ロイヤルティプログラム参加率 (Loyalty Program Participation Rate)**
    *   **選定理由**: ロイヤルティプログラムへの参加は、顧客のブランドへのコミットメントと将来的な継続的な購買意欲を示す強い指標です。参加顧客は非参加顧客と比較してCLTVが高い傾向にあります。
    *   **測定可能列**: `Loyalty_Program_Member`

4.  **顧客満足度 (Customer Satisfaction Score)**
    *   **選定理由**: 高い顧客満足度は、リピート購入、ポジティブな口コミ、そして顧客の長期的な維持に不可欠です。満足度の向上は離反率の低下とCLTVの増加に直結します。
    *   **測定可能列**: `Customer_Rating`

5.  **製品返品率 (Product Return Rate)**
    *   **選定理由**: 高い製品返品率は、顧客の不満、製品品質の問題、または期待とのミスマッチを示唆し、将来の購買意欲やロイヤルティを低下させる可能性があります。低い返品率は顧客満足度とCLTVの維持・向上に貢献します。
    *   **測定可能列**: `Product_Returns`

## KPIツリーの構造化

### Subtask:
定義されたKGIとKPI間の因果関係を示し、ビジネスにおける意思決定ポイントを明確にするためのKPIツリーの構造を説明します。ツリーは、上位目標から下位指標への分解を明確に示します。


### KPIツリーの構造化

KGI（最終目標）である「**顧客生涯価値（CLTV）の最大化**」を頂点とし、それを構成する主要なKPI群を下位に配置します。各KPIはデータセットの特定の列によって測定され、CLTVにどのように貢献するかが示されます。

**1. KGI: 顧客生涯価値 (CLTV) の最大化**
   *   **説明**: 顧客が企業にもたらす生涯にわたる総利益の最大化を目指します。これは顧客ロイヤルティと長期的な売上収益の向上に直結します。

**2. 主要KPI (CLTVに直接影響を与える要素)**

   *   **平均注文額 (AOV)**
       *   **CLTVへの貢献**: 1回あたりの購入金額が増加することで、総収益が増加し、CLTVが向上します。
       *   **構成要素（データ列）**: `Average_Order_Value` (または `Total_Purchase_Amount` / `Frequency_of_Purchases`)
   *   **購入頻度 (Purchase Frequency)**
       *   **CLTVへの貢献**: 顧客がより頻繁に購入することで、生涯にわたる購入回数が増え、CLTVが向上します。
       *   **構成要素（データ列）**: `Frequency_of_Purchases`
   *   **ロイヤルティプログラム参加率 (Loyalty Program Participation Rate)**
       *   **CLTVへの貢献**: プログラム参加者はブランドへのコミットメントが高く、継続的な購買が見込まれるため、CLTVが高くなります。
       *   **構成要素（データ列）**: `Loyalty_Program_Member`
   *   **顧客満足度 (Customer Satisfaction Score)**
       *   **CLTVへの貢献**: 満足度が高い顧客はリピート購入しやすく、ポジティブな口コミを通じて新規顧客獲得にも貢献するため、CLTVの向上に繋がります。
       *   **構成要素（データ列）**: `Customer_Rating`
   *   **製品返品率 (Product Return Rate)**
       *   **CLTVへの貢献**: 返品率が低いことは顧客満足度が高く、製品への不満が少ないことを示し、長期的な関係維持とCLTV向上に寄与します。
       *   **構成要素（データ列）**: `Product_Returns`

**KPI間の因果関係と相互作用:**

*   **顧客満足度**の向上は、**購入頻度**の増加と**製品返品率**の低下に繋がり、ひいては**AOV**の増加（より高価な商品や追加購入）にも影響を与える可能性があります。
*   **ロイヤルティプログラム参加率**の向上は、顧客エンゲージメントを高め、**購入頻度**と**AOV**の増加に貢献することが期待されます。
*   これらのKPIが複合的に作用し、最終的に**CLTVの最大化**というKGIの達成を目指します。

このKPIツリー構造は、各KPIの動向を監視することで、どの領域に改善の余地があるかを特定し、具体的なデータ駆動型の意思決定（例：ロイヤルティプログラムの改善、製品品質向上策、顧客サービス強化など）を行うためのフレームワークを提供します。

**Reasoning**:
I will handle the missing values in `Engagement_with_Ads` and `Social_Media_Influence` by filling them with 'Unknown' as they are categorical columns with significant missing percentages. Then, I will describe the business impact and update the data quality score.



```python
print('--- 欠損値処理 ---')

# 欠損値がある列を特定
missing_cols = ['Engagement_with_Ads', 'Social_Media_Influence']

# 欠損値の補完
affected_rows = {} # 影響件数を格納する辞書
for col in missing_cols:
    if df[col].isnull().any():
        initial_missing_count = df[col].isnull().sum()
        df[col] = df[col].fillna('不明')
        final_missing_count = df[col].isnull().sum()
        if initial_missing_count > 0 and final_missing_count == 0:
            affected_rows[col] = initial_missing_count
            print(f"列 '{col}': 欠損値 {initial_missing_count} 件を '不明' で補完しました。")
        else:
            print(f"列 '{col}': 欠損値の補完は不要でした、または処理後に欠損値が残っています（残存件数: {final_missing_count}）。")

print('\n更新された欠損値情報:')
missing_percentage = df.isnull().sum() / len(df) * 100
missing_info = pd.DataFrame({
    '欠損数': df.isnull().sum(),
    '欠損率 (%)': missing_percentage
}).sort_values(by='欠損率 (%)', ascending=False)
display(missing_info)

# データ品質スコアの更新
# 欠損値の適切な補完により、データがより完全になったため、スコアを更新します。
# 各列の補完につき +5 ポイントとします。
# affected_rows に補完された列のみをカウント
score_increase = len(affected_rows) * 5
data_quality_score += score_increase

print(f'\nデータ品質スコア: {data_quality_score} (+{score_increase} ポイント)')

# ビジネス影響の説明
markdown_text = f"""
### 欠損値処理によるビジネスへの影響

1.  **`Engagement_with_Ads` (欠損値 {affected_rows.get('Engagement_with_Ads', 0)} 件を '不明' で補完)**:
    *   広告エンゲージメントに関する情報が欠損しているレコードについて、『不明』というカテゴリを割り当てることで、これらのデータを分析から除外することなく活用できるようになりました。これにより、広告の効果測定や顧客セグメンテーションにおいて、より多くのデータを考慮に入れることができ、分析の網羅性が向上します。
    *   **データ品質スコアへの影響**: 欠損値の適切な補完により、分析対象データが増加し、バイアスの可能性を低減したため、データ品質スコアを +5 ポイント増加させます。

2.  **`Social_Media_Influence` (欠損値 {affected_rows.get('Social_Media_Influence', 0)} 件を '不明' で補完)**:
    *   ソーシャルメディアの影響力に関する情報が欠損しているレコードについても、『不明』というカテゴリで補完しました。これにより、ソーシャルメディアの影響力と購買行動の関連性を分析する際に、情報が欠損している顧客層の傾向も考慮できるようになります。マーケティング戦略の策定において、より包括的な視点を提供します。
    *   **データ品質スコアへの影響**: 欠損値の適切な補完により、分析対象データが増加し、バイアスの可能性を低減したため、データ品質スコアを +5 ポイント増加させます。

**データ品質スコアの更新**: 現在のスコア {data_quality_score - score_increase} + {score_increase} (欠損値補完) = {data_quality_score}
"""

print(markdown_text)
```

I will also update the data quality score based on the outlier detection and handling.



```python
print('--- 外れ値検証（IQR法） ---')

initial_data_quality_score_outlier = data_quality_score
outlier_impact_summary = []
outliers_found_total = 0

# 数値列を特定
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()

# 除外する列 (IDなど、外れ値の対象としないもの)
exclude_cols = ['Customer_ID']
numeric_cols = [col for col in numeric_cols if col not in exclude_cols]

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # 外れ値を検出
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    num_outliers = outliers.shape[0]

    if num_outliers > 0:
        outliers_found_total += num_outliers
        outlier_impact_summary.append(f"'{col}' 列で {num_outliers} 件の外れ値が見つかりました（下限 {lower_bound:.2f}, 上限 {upper_bound:.2f}）。")

        # 外れ値の処理: 例として、上下限でクリップする（ビジネス要件に応じて変更）
        df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
        df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
        outlier_impact_summary.append(f"  -> '{col}' 列の外れ値を上下限でクリップしました。")
        data_quality_score -= 5 # 外れ値の修正でスコアを増やすのではなく、異常があったことで減点されると考える
    else:
        outlier_impact_summary.append(f"'{col}' 列に外れ値は見つかりませんでした。")

print('\n--- 外れ値処理結果 ---')
for summary in outlier_impact_summary:
    print(f'- {summary}')

print(f'\n合計外れ値件数: {outliers_found_total} 件')
print(f'更新されたデータ品質スコア: {data_quality_score} ({data_quality_score - initial_data_quality_score_outlier} ポイント)')

# ビジネス影響の説明 (Markdown形式)
markdown_text_outlier = f"""
### 外れ値検証と処理によるビジネスへの影響

**概要**: IQR法を用いて複数の数値列の外れ値の検証と処理を実施しました。これには、年齢、購入金額、購入頻度、製品リサーチ時間、顧客評価、製品評価などの列が含まれます。

**影響内容**:
"""
for summary in outlier_impact_summary:
    markdown_text_outlier += f"*   {summary}\n"

markdown_text_outlier += f"""

**ビジネスへの影響**: 外れ値の適切な処理により、統計的分析の精度が向上し、モデル構築時のバイアスが減少します。例えば、極端に高いまたは低い購入金額が補正されることで、平均購入額やCLTVの推定がより現実的になり、マーケティング戦略や予算配分に関する意思決定がより堅牢になります。外れ値によるデータの歪みが排除され、より信頼性の高い分析結果に基づいたインサイトを提供できます。

**データ品質スコアの更新**: 外れ値が検出され修正された各カテゴリについて、データ品質スコアを -5 ポイントとしました。これは、データに潜在的な問題があったことの反映です。
*   **変更前スコア**: {initial_data_quality_score_outlier}
*   **変更後スコア**: {data_quality_score}
*   **増減**: {data_quality_score - initial_data_quality_score_outlier} ポイント
"""

print(markdown_text_outlier)
```

## Prepare Time-Series Data

### Subtask:
Aggregate the 'Purchase_Amount' and 'Frequency_of_Purchase' from the 'Time_of_Purchase' column to a suitable time granularity (e.g., monthly or weekly) to prepare for time-series analysis.


**Reasoning**:
Extract the year-month from 'Time_of_Purchase' and aggregate purchase amount and frequency by month, then display the resulting time-series data.



```python
print('--- Prepare Time-Series Data ---')

# 1. Extract the 'year-month' from the Time_of_Purchase column
df['Purchase_Month'] = df['Time_of_Purchase'].dt.to_period('M')

# 2. Group by Purchase_Month and calculate the sum of Purchase_Amount and Frequency_of_Purchase
time_series_df = df.groupby('Purchase_Month').agg({
    'Purchase_Amount': 'sum',
    'Frequency_of_Purchase': 'sum'
}).reset_index()

# 4. Sort the DataFrame by Purchase_Month
time_series_df = time_series_df.sort_values(by='Purchase_Month')

# Convert Purchase_Month back to datetime for plotting later
time_series_df['Purchase_Month'] = time_series_df['Purchase_Month'].dt.to_timestamp()

# 5. Display the head of the newly created time-series DataFrame
print("Monthly aggregated time-series data:")
display(time_series_df.head())
```

## Statistical Validation

### Subtask:
Perform statistical tests (e.g., t-tests or ANOVA) to validate the significance of differences observed in group comparison analyses (e.g., KPIs by Gender, Income Level, Marital Status, Loyalty Program Membership). This will add statistical rigor to the identified insights.


**Reasoning**:
I need to determine the appropriate statistical test (t-test or ANOVA) for each combination of categorical variables and KPIs, and then perform these tests. This involves checking the number of unique categories for each categorical variable to decide which test to apply.



```python
from scipy import stats

kpis = ['Purchase_Amount', 'Frequency_of_Purchase', 'Customer_Satisfaction']
categorical_segment_vars = ['Gender', 'Income_Level', 'Marital_Status', 'Customer_Loyalty_Program_Member']

print("--- Statistical Validation of Group Differences ---")
print("Null Hypothesis (H0): There is no significant difference in KPI means/medians across groups.")
print("Alternative Hypothesis (H1): There is a significant difference in KPI means/medians across groups.")

significance_level = 0.05 # Alpha value

for cat_var in categorical_segment_vars:
    print(f"\nAnalyzing by: {cat_var}")
    unique_categories = df[cat_var].dropna().unique()

    if len(unique_categories) < 2:
        print(f"  Skipping {cat_var}: Not enough unique categories for statistical test.")
        continue

    for kpi in kpis:
        print(f"  KPI: {kpi}")

        # Prepare groups for testing
        groups = [df[df[cat_var] == category][kpi].dropna() for category in unique_categories]

        # Check if there are at least two non-empty groups to compare
        non_empty_groups = [group for group in groups if not group.empty]
        if len(non_empty_groups) < 2:
            print(f"    Skipping {kpi} for {cat_var}: Not enough data in groups for comparison.")
            continue

        if len(unique_categories) == 2: # Two groups, use independent t-test
            # Perform Levene's test for homogeneity of variances
            stat_levene, p_levene = stats.levene(*non_empty_groups)
            equal_var = p_levene > significance_level # If p_levene > alpha, assume equal variances

            # Perform independent t-test
            stat, p_value = stats.ttest_ind(*non_empty_groups, equal_var=equal_var)
            test_type = "Independent t-test"

        else: # More than two groups, use ANOVA
            # Check for sufficient variance within groups for ANOVA
            sufficient_variance = all(group.var() > 0 for group in non_empty_groups)
            if not sufficient_variance:
                print(f"    Skipping {kpi} for {cat_var}: Not enough variance within groups for ANOVA.")
                continue

            stat, p_value = stats.f_oneway(*non_empty_groups)
            test_type = "ANOVA"

        print(f"    Test: {test_type}")
        print(f"    P-value: {p_value:.4f}")

        if p_value < significance_level:
            print(f"    Result: Reject H0. There is a statistically significant difference in {kpi} across {cat_var} groups.")
        else:
            print(f"    Result: Fail to reject H0. No statistically significant difference in {kpi} across {cat_var} groups.")
```

本プロジェクトの主要な目標は、Eコマースデータから顧客行動と購買パターンを分析し、顧客生涯価値（CLTV）を最大化することです。CLTVを追跡するための主要業績評価指標（KPI）を定義し、顧客セグメンテーション、購買トレンド、ロイヤルティ、顧客満足度などの要因に焦点を当てました。データ理解、クリーニング、探索的データ分析（EDA）を通じて、顧客ロイヤルティと収益最大化に貢献するデータ駆動型の意思決定を支援するための実用的なインサイトを導き出しました。

#### 2. データ概要
**データソース**: Eコマース顧客行動分析データ
**初期データ理解**: データの読み込み後、以下の初期チェックを実施しました。
*   **行数・列数**: 1000行、28列のデータセット
*   **データ型**: `Purchase_Amount` (object) および `Time_of_Purchase` (object) が数値型と日付時刻型に変換されました。`Engagement_with_Ads` と `Social_Media_Influence` には欠損値が存在しました。
*   **欠損値処理**: `Engagement_with_Ads` と `Social_Media_Influence` の欠損値は「不明」で補完され、それぞれ256件、247件が影響を受けました。これにより、分析対象データが増加し、バイアスの可能性が低減されました。
*   **重複行**: 重複行は検出されませんでした。
*   **異常値（業務ロジック）**: `Age`, `Purchase_Amount`, `Frequency_of_Purchase`, `Time_Spent_on_Product_Research(hours)`, `Customer_Satisfaction`, `Product_Rating` の各列において業務ロジックに基づく異常値チェックを実施しましたが、異常値は検出されませんでした。
*   **外れ値（IQR法）**: `Age`, `Purchase_Amount`, `Frequency_of_Purchase`, `Brand_Loyalty`, `Product_Rating`, `Time_Spent_on_Product_Research(hours)`, `Return_Rate`, `Customer_Satisfaction`, `Time_to_Decision` の各数値列でIQR法を用いた外れ値検出を行いましたが、外れ値は検出されませんでした。
**最終データ品質スコア**: 初期スコア100から、データ型変換で+10、欠損値処理で+10され、最終的に **120** となりました。業務ロジックと外れ値検証では異常・外れ値がなかったためスコアの変動はありませんでした。

#### 3. 分析手法
本分析では、以下の手法を適用しました。
*   **探索的データ分析 (EDA)**: 主要な数値変数とカテゴリ変数の分布分析。
*   **グループ比較分析**: `Gender`, `Income_Level`, `Marital_Status`, `Customer_Loyalty_Program_Member` などのカテゴリ変数に基づくKPI (`Purchase_Amount`, `Frequency_of_Purchase`, `Customer_Satisfaction`) の比較。
*   **貢献度分析**: `Purchase_Category`, `Location`, `Purchase_Channel` ごとの総購買額の集計。
*   **時系列分析**: `Purchase_Amount` と `Frequency_of_Purchase` の月次トレンド分析。
*   **統計的検証**: グループ比較で観察された差の統計的有意性を評価するためのt検定またはANOVA。
*   **Impact x Effort 分析**: 提案された推奨事項の実行優先順位付け。

#### 4. 主要なインサイト
**a. 初期EDAの発見**:
*   **数値変数**: `Age`, `Purchase_Amount`, `Frequency_of_Purchase` は比較的均一な分布を示しました。`Time_Spent_on_Product_Research(hours)` と `Return_Rate` は二峰性の分布を示し、異なる顧客セグメントの存在を示唆します。`Customer_Satisfaction` は高評価に偏っており、`Product_Rating` と `Time_to_Decision` は多様な分布を示しました。
*   **カテゴリ変数**: `Gender` はほぼ均等、`Income_Level` は中高所得層が多数、`Marital_Status` は既婚と独身が多数を占めます。`Loyalty_Program_Member`、`Purchase_Channel`、`Purchase_Category` の分布は、ターゲットマーケティングやチャネル最適化のための基礎情報を提供しました。

**b. グループ比較の結果**:
*   **Gender**: `Polygender` の顧客が平均購入額と購入頻度で高い傾向を示しました。統計的検証では、性別によるKPIの平均値に有意な差は見られませんでした (p > 0.05)。
*   **Income_Level**: 収入レベル間でKPIに大きな差は見られませんでした。統計的検証でも有意な差は検出されませんでした (p > 0.05)。
*   **Marital_Status**: 婚姻状況間でKPIに大きな差は見られませんでした。統計的検証でも有意な差は検出されませんでした (p > 0.05)。
*   **Customer_Loyalty_Program_Member**: ロイヤルティプログラム会員は非会員よりも平均 `Purchase_Amount` が統計的に有意に高い傾向がありました (p = 0.0011)。ただし、`Frequency_of_Purchase` と `Customer_Satisfaction` には有意な差は見られませんでした (p > 0.05)。

**c. 貢献度分析の結果**:

*   **ビジネス示唆**: ピーク時にはプロモーション強化と在庫確保、オフピーク時には顧客の再活性化施策が有効です。

*   **サービス品質の定量的評価**: 顧客サービスに関する詳細な定量的指標が不足しており、顧客満足度への影響を詳細に分析することは困難です。

*   **統計的検証**: グループ比較で観察された差の統計的有意性を評価するためのt検定またはANOVA。
*   **Impact x Effort 分析**: 提案された推奨事項の実行優先順位付け。

#### 4. 主要なインサイト
**a. 初期EDAの発見**:
*   **数値変数**: `Age`, `Purchase_Amount`, `Frequency_of_Purchase` は比較的均一な分布を示しました。`Time_Spent_on_Product_Research(hours)` と `Return_Rate` は二峰性の分布を示し、異なる顧客セグメントの存在を示唆します。`Customer_Satisfaction` は高評価に偏っており、`Product_Rating` と `Time_to_Decision` は多様な分布を示しました。

**b. グループ比較の結果**:
*   **Gender**: `Polygender` の顧客が平均購入額と購入頻度で高い傾向を示しました。統計的検証では、性別によるKPIの平均値に有意な差は見られませんでした (p > 0.05)。
*   **Income_Level**: 収入レベル間でKPIに大きな差は見られませんでした。統計的検証でも有意な差は検出されませんでした (p > 0.05)。
*   **Marital_Status**: 婚姻状況間でKPIに大きな差は見られませんでした。統計的検証でも有意な差は検出されませんでした (p > 0.05)。
*   **Customer_Loyalty_Program_Member**: ロイヤルティプログラム会員は非会員よりも平均 `Purchase_Amount` が統計的に有意に高い傾向がありました (p = 0.0011)。ただし、`Frequency_of_Purchase` と `Customer_Satisfaction` には有意な差は見られませんでした (p > 0.05)。

**c. 貢献度分析の結果**:
*   **Purchase_Category**: `Jewelry & Accessories`, `Sports & Outdoors`, `Electronics` が上位の購買カテゴリであり、総購買額の大部分を占めていました。

*   **Purchase_Channel**: `Mixed`, `Online`, `In-Store` のチャネル間で総購買額の差は比較的小さく、いずれのチャネルも重要な貢献をしていました。

**d. 時系列トレンドの結果**:
*   **Purchase_Amount**: 月間総購買金額は3-4月と7-8月にピークを持つ季節性を示し、5月と12月には落ち込みが見られました。
*   **Frequency_of_Purchase**: 月間総購入頻度も購買金額と同様の季節性を示し、8月にピーク、12月に最低を記録しました。
*   **ビジネス示唆**: ピーク時にはプロモーション強化と在庫確保、オフピーク時には顧客の再活性化施策が有効です。

#### 6. ROI分析の実現可能性と代替案
**実現可能性**: 現在のデータセットには、製品の仕入れコスト、製造コスト、販売関連コスト、顧客獲得コスト（CAC）など、本格的なROI分析に必要な直接的なコストデータや単価データが存在しません。
**代替案**:
*   **データ取得**: 会計システムやERP、製品データベース、CRM/SFAからのデータ統合を推奨。
*   **代替指標**: 粗利益率の近似、平均注文額（AOV）、顧客生涯価値（CLTV）、返品率（`Return_Rate`）の最適化を収益性評価の代替指標として活用することを提案します。

#### 7. 制限事項
*   **CLTVの直接的な算出**: 将来予測モデルの構築には、時系列詳細データやCACなどの収益性情報が不足しています。
*   **キャンペーン効果測定**: キャンペーン識別子や実施期間に関する情報がないため、マーケティング活動の直接的な効果測定はできません。
*   **競合データ**: 業界内での相対的なパフォーマンス評価に必要な競合データは含まれていません。
*   **サービス品質の定量的評価**: 顧客サービスに関する詳細な定量的指標が不足しており、顧客満足度への影響を詳細に分析することは困難です。

これらの制限事項は、将来のデータ収集戦略や追加分析の方向性を決定する上で重要です。
